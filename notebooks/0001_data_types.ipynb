```python
# Native Data Types in Python

a = "Tilak Bhatt" # string
b = 1234 # number integer
c = None # NoneType
d = 3.14 # number float
e = True # boolean

# printing the regular way 
print(a)
print(b)
print(c)
print(d)
print(e)

print("\n")

# printing using f-string
print(f"String: {a}")
print(f"Integer: {b}")
print(f"NoneType: {c}")
print(f"Float: {d}")
print(f"Boolean: {e}")

print("\n")

# type of 

print(f"Type of a: {type(a)}")
print(f"Type of b: {type(b)}")
print(f"Type of c: {type(c)}")
print(f"Type of d: {type(d)}")
print(f"Type of e: {type(e)}")
```

### 1. List
Think of list as array

```python
# list is basically array
# could have different data types in the same list

my_list = ["Tilak", "Bob", "Alice", 1, 2, 3, None, True, 3.14]

# 1. indexing
print(my_list[0]) # zeroth
print(my_list[-1]) # last

print("\n")

# 2. slicing
print(my_list[:3]) # from zeroth to index 2
print(my_list[2:]) # from index 2 to end
print(my_list[1:4]) # from index 1 to index 3
print(my_list[::2]) # from start to end with step 2
print(my_list[::-1]) # from end to start with step -1
print(my_list[1:7:2]) # from index 1 to index 6 with step 2

# 3. appending
print("\n")
my_list.append("Charlie")
print(my_list)

# 4. removing
my_list.remove("Bob")
print(my_list)

# 5. length
print("\n")
print(len(my_list))

# 6. extending
print("\n")
my_list.extend(["Dave", "Eve"]) # ADDS FLATTENED
print(my_list)


# 7. add at beginning
print("\n")
my_list.insert(0, "Zara")
print(my_list)

# 8. popping
print("\n")
popped_element = my_list.pop() # removes last element and returns it
print(f"Popped element: {popped_element}")
print(my_list)

# 9. insert at index 2
print("\n")
my_list.insert(2, "Frank")
print(my_list)
```

```python
# tuple 
# similar to list but immutable

my_tuple = ("Tilak", "Bob", "Alice", 1, 2, 3, None, True, 3.14)

# empty tuple
empty_tuple = ()

# you cannot change/modify elements
# my_tuple[0] = "Charlie" # This will raise an error

# you cannot append but concatinate

new_tuple = empty_tuple + (1, 2)
new_tuple

# tuple into list 
tuple_to_list = list(my_tuple)
tuple_to_list
```

```python
# set
# collection of unique elements
# unordered, unindexed, mutable

my_set = {"Tilak", "Bob", "Alice", 1, 2, 2, 3, None, True, 3.14, (1,2,3), (1,2,3)} # duplicate elements will be removed

# empty set
empty_set = set()

# get element
# you cannot get element by index because it's unordered and unindexed
# but you can check if an element is in the set
print("Tilak" in my_set) # True
print("Charlie" in my_set) # False
print((1,2,3) in my_set) # True

# add element 
my_set.add("jj")
print(my_set)

my_set.update(["kk", "ll"]) # adds multiple elements
print(my_set)
```

```python
# dictionary
# think of this as js object not quite but key value pairs

my_dict = {
    "name": "Tilak Bhatt",
    "age": 30,
    "is_student": False,
    "hobbies": ["coding", "gaming", "traveling"],
    "address": {
        "street": "123 Main St",
        "city": "New York",
    }
}

# get value by key
print(my_dict["name"]) # Tilak Bhatt
print(my_dict["address"]["city"]) # New York

# add key value pair
my_dict["email"] = "tilak@domain.com"
my_dict
# add neseted key value pair
my_dict["address"]["zip"] = "10001"
my_dict

# update value
my_dict["age"] = 31
```
