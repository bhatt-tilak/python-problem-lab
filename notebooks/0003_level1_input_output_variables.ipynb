### 5. Take two numbers and print all arithmetic operations

```python
a = 20
b = 5

print(a + b) # sum
print (a - b) # difference
print (a * b) # product
print (a / b) # quotient, division always returns a float in Python 3
print (a // b) # floor division, returns the quotient without the decimal part
print (a % b) # modulus, returns remainder
print (a ** b) # exponential

print (a**2 + b**2) # sum of squares
```

### 6. Convert Celsius to Fahrenheit

```python
celcius = 100
print(9/5)
fah = (celcius * 9/5) + 32
print("Farenheit", fah)
```

### 7. Convert Fahrenheit to Celsius

```python
fah = 212

celcius = (fah - 32) * 5/9
print("Celcius", celcius)
```

### 8. Calculate area of a rectangle

```python
lenght = 10
breadth = 5
# area = length * breadth
area = lenght * breadth
print(f"Area {area} sq units")
```

### 9. Calculate area of a circle

```python
radius = 5
pi = 3.1415

area = pi * radius ** 2
print(f"Area: {area} sq units")
```

### 10. Swap two variables without using a third variable

```python
a = 10
b = 20
c = a + b
print (f"{a} {b}")

b = a
a = c - a
print(f"{a} {b}")
```
